```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id="use-your-client-id", client_secret="use-your-client-secret")
```

```python
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
playlist_link = "https://open.spotify.com/playlist/37i9dQZF1DXcBWIGoYBM5M"
```

```python
playlist_link.split('/')
```

```python
playlist_URI = playlist_link.split('/')[-1]
```

```python
data = sp.playlist_tracks(playlist_URI)
```

## Album Dataframe

```python
album_list = []
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    releasedate = row['track']['album']['release_date']
    total_tracks = row['track']['album']['total_tracks']
    url = row['track']['album']['external_urls']['spotify']
    album_dict = {'album_id': album_id, 'album_name': album_name, 'release_date': releasedate, 'total_tracks': total_tracks, 'url': url}
    album_list.append(album_dict)
album_list
```

```python
album_df = pd.DataFrame.from_dict(album_list)
album_df.head()
```

```python
album_df['release_date'] = pd.to_datetime(album_df['release_date'])
```

```python
album_df.info()
```

```python
album_df = album_df.drop_duplicates(subset=['album_id'])
```

```python
album_df.to_excel('./album.xlsx')
```

## Artist Dataframe

```python
artist_list = []
for row in data['items']:
    for key, value in row.items():
        if(key == "track"):
            for artist in value['artists']:
                artist_dict = {'artist_id': artist['id'], 'artist_name':artist['name'], 'external_url': artist['href']}
                artist_list.append(artist_dict)
```

```python
artist_df = pd.DataFrame.from_dict(artist_list)
artist_df.head()
```

```python
artist_df = artist_df.drop_duplicates(subset=['artist_id'])
```

## Song Dataframe

```python
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_dict = {'song_id':song_id,'song_name':song_name,'duration_ms':song_duration,'url':song_url,
                    'popularity':song_popularity,'song_added':song_added,'album_id':album_id,
                    'artist_id':artist_id
                   }
    song_list.append(song_dict)
```

```python
song_df = pd.DataFrame.from_dict(song_list)
```

```python
song_df.head()
```

```python
song_df = song_df.drop_duplicates(subset=['song_id'])
```
